### For the top-selling game of all time, how many standard deviations above/below the mean are its sales for North America?

```python
df = pd.read_csv('vgsales.csv')

sorted_by_top_game = df.sort_values('Global_Sales', ascending=False)
```

```python
sorted_by_top_game[:1]
```

### how many standard deviations above/below the mean are its sales for North America?

```python
mean_sales = np.mean(sorted_by_top_game)
```

```python
mean_sales = mean_sales['Global_Sales']
```

```python
mean_sales
```

```python
deviation_sales = np.std(sorted_by_top_game)
```

```python
deviation_sales = deviation_sales['Global_Sales']

```

```python
deviation_sales
```

```python
top_game_sales = sorted_by_top_game.loc[0]['Global_Sales']
```

```python
top_game_sales
```

```python
result = ((top_game_sales - mean_sales) / deviation_sales)
```

```python
result
```

#### so there is about 52-53 standard deviations above the mean 

## The Nintendo Wii seems to have outdone itself with games. How does its average number of sales compare with all of the other platforms?


#### first solution


```python


wii_games = df[df.Platform == 'Wii']
not_wii_games = df[df.Platform != 'Wii']
wii_game_average = wii_games.Global_Sales.mean()
not_wii_game_average = not_wii_games.Global_Sales.mean()
print('Wii avg: ' + str(wii_game_average))
print('Not-Wii avg: ' + str(not_wii_game_average))

```

#### wii average price compare to all game average price:


```python
results = wii_game_average - not_wii_game_average
```

```python
results
```
